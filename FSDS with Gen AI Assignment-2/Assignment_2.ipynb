1. Email Validator (String Filtering + Loops)
üíº Real-world use: Used in signup forms or data cleaning tasks.
Task:
Check if a given email is valid ‚Äî must contain ‚Äú@‚Äù and end with ‚Äú.com‚Äù or ‚Äú.org‚Äù.
Example:
Input: hello@inceptionbd.com
Output: Valid Email

```python
# 1
email = input("enter email: ")                      
has_email = "@" in email
last_four= email[-4:]
# print(last_four)
pos_end = last_four == ".com" or last_four == ".org"  

if has_email and pos_end:
    print('Valid Email')
else:
    print("invalid")     
```

3. Password Strength Checker
üíº Real-world use: Used in user authentication systems.
Task:
Check if a password is strong ‚Äî must have at least one uppercase, one lowercase, one digit, and one special character.
Example:
Input: "Hello@123"
Output: Strong Password

```python

# 3
a= input('enter email:')
has_upper= False
has_lower= False
has_digit= False
has_special=False

for i in a:
    if i.isupper():
        # print(i.isupper)
        has_upper = True
    elif i.islower():
        has_lower = True
    elif i.isdigit():
        has_digit=True
    else:
        has_special=True
if has_upper and has_lower and has_digit and has_special:
    print('strong')
else:
    print('not strong')
```

4. Username Generator
üíº Real-world use: Automatically create usernames for users.
Task:
Take user‚Äôs full name and generate a short username.
Rules: first 3 letters of first name + last 3 letters of last name + random number (use loops only).
Example:
Input: "Mamun Malitha"
Output: mamtha57

```python
# 4

full_name = input("Enter full name: ")
parts = full_name.split()
print(parts)
first = parts[0]
last = parts[-1]

username = first[:3] + last[-3:]

num = 0
for i in range(len(full_name)):
    num = (num * 7 + i * 3 + len(full_name)) % 100
num = num % 100

print(f"{username.lower()}{num}")
```

6. Remove Duplicates from Contact List
üíº Real-world use: Data cleaning in CRM or user databases.
Task:
Given a list of names (comma separated), remove duplicates.
Example:
Input: "Rahim, Karim, Rahim, Sultana"
Output: "Rahim, Karim, Sultana"

```python
# 6
s=input("enter:")

parts = s.split(",")
c=[]
for n in parts:
    c.append(n.strip())

seen =[]
for item in c:
    if item not in seen:
        seen.append(item)
print(", ".join(seen))
        
        
```

10. Word Reversal for Data Encryption
üíº Real-world use: Simple text obfuscation or playful encoding.
Task:
Reverse each word in a sentence while keeping word order the same.
Example:
Input: "AI is fun"
Output: "IA si nuf"

```python
# 10
a= input('AI is fun')
b=a.split()
# print(b)
for i in b:
    c= i[::-1]
    print(c,end=' ')                            
```
