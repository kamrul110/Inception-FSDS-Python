7. Discount Calculator
Real-world use: Online store discount engine.
Task:
Write a function apply_discounts(products) where:
 products is a list of tuples: (name, price, discount)
 return a list of updated prices after discount
Example:
Input (List of Tuples): [("Keyboard", 80.00, 20), ("Mouse", 25.00, 10), ("Monitor", 300.00, 5)]
Output (List of Prices): [64.00, 22.50, 285.00]

```python
def product_dis(value):
    # print(e)
    update_price=[]
    for item in value:
        name = item[0]
        price = float(item[1])
        discount= float(item[2])
        
        dis_ammount = price*(discount/100)
        final_price = price- dis_ammount
        update_price.append(final_price)
    return update_price
    
user_Input = input('enter: ')
b= user_Input.split(',')
c= []
for i in b:
    name,price,discount= i.split(':')
    c.append((name,float(price),float(discount)))
print("Input (List of Tuples):",c)   
result= product_dis(c)
print("Output (List of Prices):",result)
```
